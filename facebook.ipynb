```python
import pandas as pd
```

```python
df = pd.read_csv("dataset_Facebook.csv", sep = ";")
df
```

```python
df.head()
```

```python
df.shape
```

```python
# Create Subset
d1 = df.loc[1:3]
d2 = df.loc[4:6]
```

```python
# Merge Data
merge1 = pd.merge(d1,d2, how = "outer")
merge1
```

```python
merge2 = pd.merge(d1,d2, how = "left") # left will take all common and elements from left dataframe
merge2
```

```python
merge3 = pd.merge(d1, d2, how = "right")
merge3
```

```python
merge4 = pd.merge(d1, d2, on = "Type", how = "inner")
merge4
```

```python
merge5 = pd.merge(d1, d2,  how = "cross")
merge5
```

```python
# Sorting 
d3 = df.head(5)
d3.sort_values(by = "Category")
d3
```

```python
# 4.Transpose Operation
d3.transpose()
```

```python
# 5. Shape Reshape Operation
d3.shape
d3.melt(id_vars = ["Type"])
d3.pivot(columns = ["Type"])
```
