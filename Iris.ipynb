```python
# Distplot shows the overall distribution of continuous data
sb.set_style('whitegrid')
sb.distplot(df['PetalLengthCm'],color ='red', kde=False)
```

```python
sb.histplot(df['PetalWidthCm'],kde=True, bins=5, fill=False, stat='probability')  # kde = kernel density estimate(It is a curve line over graph)
```

```python
sb.set_style('ticks')
sb.distplot(df['PetalLengthCm'])
```

```python
#Displays default color palette.
sb.color_palette()
```

```python
# Scatter plot is plot with the dots.
sb.scatterplot(x= df['PetalWidthCm'], y=df['Species'], palette='terrain')
```

```python
sb.lineplot(x='SepalLengthCm', y='SepalWidthCm', data=df, color='blue')
#  OR
# sb.lineplot(x=df['SepalLengthCm'], y=df['SepalWidthCm'], color='blue')
```

```python
sb.set_style('white')
sb.FacetGrid(df)
```

```python
g= sb.FacetGrid(df, col='Species')
g.map(sb.histplot, "PetalLengthCm", color='blue')
```

```python
g= sb.FacetGrid(df, col='Species')
g.map_dataframe(sb.histplot, y="PetalLengthCm", color='blue')
```
