```python
# Model Building 

from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
dfm = df.loc[0:20,['Date','NO2(GT)','PT08.S5(O3)']]
dfm
```

```python
# Split into input and output features

inputs = dfm[["NO2(GT)","PT08.S5(O3)"]]
target = dfm["Date"]
```

```python
target
```

```python
inputs_train, inputs_test, target_train, target_test = train_test_split(inputs, target, test_size = 0.3)
```

```python
target_test
```

```python
model = DecisionTreeClassifier()
model.fit(inputs_train, target_train)
```

```python
model.score(inputs_train, target_train)
```

```python
target_test
```

```python
model.predict([[113.0,1268.0]])
```

```python
predicted = model.predict(inputs_test)
predicted
```

```python
model.score(inputs_test, predicted)
```

```python
dfm1 = inputs_test.copy()
dfm1['actual'] = target_test
dfm1['predicton'] = predicted
dfm1
```
